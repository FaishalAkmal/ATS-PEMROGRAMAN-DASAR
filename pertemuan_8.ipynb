```python
Faishal_akmal['jurusan'] = Faishal_akmal['jurusan'].str.lower().str.strip()
Faishal_akmal.jurusan.value_counts()
```

```python
mapping_jurusan = {
    "energi": "Pembangkit Energi",
    " energi ": "Pembangkit Energi",
    "ENERGI": "Pembangkit Energi",
    "EnErGi": "Pembangkit Energi",
    "eng": "Pembangkit Energi",
    "teknik": "Pembangkit Energi",
    "Teknik ": "Pembangkit Energi",
    "teknik ": "Pembangkit Energi",
    "manajemenenergi": "Manajemen Energi",
    "manajemen energi": "Manajemen Energi"
}
Faishal_akmal["jurusan"] = Faishal_akmal["jurusan"].replace(mapping_jurusan)
Faishal_akmal["jurusan"] = Faishal_akmal["jurusan"].str.title()
Faishal_akmal.jurusan.value_counts()
```
