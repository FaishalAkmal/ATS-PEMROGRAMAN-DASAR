2. FUNCTIONS

```python
def hitung_efisiensi(input_energi, output_energi):
    efisiensi = output_energi / input_energi
    return efisiensi

hasil = hitung_efisiensi(500, 400)
print("Efisiensi turbin =", hasil)
```

Fungsi hitung_efisiensi() memiliki dua parameter, yaitu input_energi dan output_energi. Di dalam fungsi ini, efisiensi dihitung menggunakan rumus output / input untuk mengetahui seberapa besar energi keluaran dibandingkan dengan energi masukan. Nilai hasil perhitungan tersebut kemudian dikembalikan (return) kepada pemanggil fungsi. Pada bagian bonus, fungsi ini dipanggil dengan nilai input_energi = 500 dan output_energi = 400, sehingga diperoleh hasil efisiensi sebesar 0.8 atau setara dengan 80%, yang menunjukkan bahwa turbin bekerja dengan efisiensi 80%.

4. LOGIKA IF MAJEMUK/BERTINGKAT

```python
tekanan_uap = 180.0
status_manual_override = False

if tekanan_uap > 150 and status_manual_override == False:
    print("Katup Terbuka")
else:
    print("Katup Tertutup")
```

Program ini memeriksa dua kondisi sekaligus menggunakan operator logika and. Kondisi pertama adalah tekanan_uap 180, yang berarti tekanan harus lebih besar dari 180 psi. Kondisi kedua adalah status_manual_override == False, yang menunjukkan bahwa katup tidak boleh dibuka secara manual. Jika kedua kondisi tersebut terpenuhi, maka program akan mencetak "Katup Terbuka". Namun, jika salah satu dari kondisi tersebut tidak terpenuhi, maka program akan mencetak "Katup Tertutup".

6. VARIABLE DAN EKSPRESI

```python
tegangan = 220
arus = 15

daya = tegangan * arus

print("Tegangan =", tegangan, "Volt")
print("Arus =", arus, "Ampere")
print("Daya listrik =", daya, "Watt")
```

Hasil perhitungan ini kemudian ditampilkan ke layar menggunakan fungsi print().

8. FUNGSI / MODULARISASI

```python
def hitung_efisiensi_termal(energi_keluar, energi_masuk):
    efisiensi = (energi_keluar / energi_masuk) * 100
    print("Efisiensi termal turbin gas =", efisiensi, "%")

hitung_efisiensi_termal(400, 500)
```

Fungsi hitung_efisiensi_termal() memiliki dua parameter, yaitu energi_keluar sebagai output energi dan energi_masuk sebagai input energi. Di dalam fungsi, efisiensi dihitung menggunakan rumus Efisiensi = (energi_keluar / energi_masuk) Ã— 100. Nilai efisiensi yang diperoleh kemudian ditampilkan menggunakan fungsi print(). Saat fungsi dipanggil dengan nilai energi_keluar = 400 dan energi_masuk = 500, maka hasil perhitungannya adalah Efisiensi = (400 / 500) Ã— 100 = 80%, yang berarti efisiensi termal turbin gas tersebut sebesar 80%.

10. WHILE LOOP AND LOGIKA

```python
tangki = 0

while tangki <= 100:
    print("Volume tangki:", tangki, "liter")

    if tangki == 80:
        print("WARNING: Hampir penuh!")
    elif tangki == 100:
        print("TANGKI PENUH - Pengisian berhenti.")
        break

    tangki += 10
```

Variabel tangki digunakan untuk menyimpan jumlah bahan bakar saat ini yang dimulai dari 0 liter. Perulangan while tangki <= 100 berarti program akan terus berjalan selama kapasitas tangki belum melebihi 100 liter. Pada setiap iterasi, volume tangki akan bertambah sebanyak 10 liter melalui perintah tangki += 10. Ketika volume tangki mencapai 80 liter, program akan menampilkan peringatan â€œWARNING: Hampir penuh!â€ sebagai tanda bahwa kapasitas hampir maksimal. Selanjutnya, jika volume tangki mencapai 100 liter, program akan menampilkan pesan â€œTANGKI PENUH â€“ Pengisian berhenti.â€ dan menghentikan proses perulangan menggunakan perintah break.

```python
def hitung_waktu_pengisian(kapasitas_baterai, daya_charger):
    """Menghitung waktu pengisian (jam)"""
    waktu = kapasitas_baterai / daya_charger
    return waktu

def tampilkan_saran(daya_charger):
    """Memberikan saran efisiensi berdasarkan daya charger"""
    if daya_charger < 5:
        return "Pengisian lambat tapi hemat energi "
    elif 5 <= daya_charger <= 10:
        return "Pengisian normal dan efisien "
    else:
        return "Pengisian cepat namun boros daya "

print("=== SIMULASI PENGISIAN BATERAI MOBIL LISTRIK ===")

while True:
    try:
        kapasitas = float(input("\nMasukkan kapasitas baterai mobil (kWh): "))
        daya = float(input("Masukkan daya charger (kW): "))

        waktu = hitung_waktu_pengisian(kapasitas, daya)
        print(f"\n Estimasi waktu pengisian: {waktu:.2f} jam")
        print(tampilkan_saran(daya))

        print("\nProgres pengisian baterai:")
        for persen in range(0, 101, 20):
            print(f" {persen}% terisi")
        print("Pengisian selesai!\n")

        ulang = input("Ingin simulasi ulang? (y/n): ")
        if ulang.lower() != 'y':
            print("Terima kasih telah menggunakan program ini!")
            break
    except ValueError:
        print("Input tidak valid! Masukkan angka yang benar.")
```

Program ini menggunakan dua fungsi utama, yaitu hitung_waktu_pengisian() dan tampilkan_saran(). Fungsi hitung_waktu_pengisian() berfungsi untuk menghitung estimasi waktu yang dibutuhkan untuk mengisi baterai mobil listrik hingga penuh berdasarkan kapasitas baterai dan daya charger yang digunakan. Sementara itu, fungsi tampilkan_saran() memberikan rekomendasi efisiensi energi berdasarkan besar daya charger yang dimasukkan oleh pengguna. Struktur percabangan if, elif, dan else digunakan untuk menentukan kategori efisiensi energi, seperti apakah pengisian tergolong lambat, normal, atau cepat. Perulangan for digunakan untuk menampilkan progres pengisian baterai secara bertahap dari 0% hingga 100%, memberikan efek simulasi proses pengisian. Selain itu, perulangan while memungkinkan pengguna untuk menjalankan simulasi lebih dari sekali tanpa harus memulai ulang program. Program ini juga dilengkapi dengan penanganan kesalahan menggunakan try dan except, yang memastikan bahwa input yang dimasukkan berupa angka sehingga program tidak mengalami error ketika dijalankan.
