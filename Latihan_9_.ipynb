```python
import pandas as pd

# Membaca dataset products.csv
products_df = pd.read_csv('products.csv')

# Menampilkan isi DataFrame
products_df
```

```python
# Ringkasan statistik
products_df.describe()

# Rata-rata harga & rata-rata quantity
mean_price = products_df['price'].mean()
mean_quantity = products_df['quantity'].mean()

print("Rata-rata harga:", mean_price)
print("Rata-rata quantity:", mean_quantity)

# Modus tipe produk
mode_product_type = products_df['product_type'].mode()
print("Modus product type:", mode_product_type)
```

```python
import matplotlib.pyplot as plt

# Distribusi Harga
products_df['price'].hist()
plt.title("Distribusi Harga Produk")
plt.xlabel("Harga")
plt.ylabel("harga terbanyak")
plt.show()

# Distribusi Quantity
products_df['quantity'].hist()
plt.title("Distribusi Jumlah (Quantity)")
plt.xlabel("Quantity")
plt.ylabel("stok")
plt.show()
```

```python
# Mean product
group_mean_type = products_df.groupby("product_type").mean(numeric_only=True)
group_mean_type

group_agg_type = products_df.groupby("product_type").agg({
    'price': ['mean', 'min', 'max'],
    'quantity': ['mean', 'min', 'max']
})
group_agg_type
```

```python
# Korelasi
corr_pq = products_df[['price', 'quantity']].corr()
print(corr_pq)

# Kovarian
cov_pq = products_df[['price', 'quantity']].cov()
print(cov_pq)

```
